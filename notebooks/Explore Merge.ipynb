```python
# --------------------------------------------------------------------Merge entre les dataframes df_incidents et df_mobilisations

# Importation de la bibliothÃ¨que Pandas pour la manipulation des donnÃ©es
import pandas as pd

# Ajuster les paramÃ¨tres pour afficher toutes les lignes et colonnes
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)

# Charger les donnÃ©es incidents et mobilisations
df_incidents = pd.read_csv("../data/Incidents.csv", sep=";", index_col=0, low_memory=False)
df_mobilisations = pd.read_csv("../data/Mobilisation.csv", sep=";", index_col=0,low_memory=False)

# RÃ©sumÃ© des informations sur chaque DataFrame 
display(df_incidents.head())
display(df_incidents.info())
display(df_incidents.describe())

display(df_mobilisations.head())
display(df_mobilisations.info())
display(df_mobilisations.describe())
```

```python

#Conversion des colonnes de dates et heures au format datetime

df_incidents.DateOfCall = pd.to_datetime(df_incidents.DateOfCall)
df_incidents.TimeOfCall = pd.to_datetime(df_incidents.TimeOfCall)
df_mobilisations.DateAndTimeMobilised = pd.to_datetime(df_mobilisations.DateAndTimeMobilised)
df_mobilisations.DateAndTimeMobile = pd.to_datetime(df_mobilisations.DateAndTimeMobile)
df_mobilisations.DateAndTimeArrived = pd.to_datetime(df_mobilisations.DateAndTimeArrived)
df_mobilisations.DateAndTimeLeft = pd.to_datetime(df_mobilisations.DateAndTimeLeft)
df_mobilisations.DateAndTimeReturned = pd.to_datetime(df_mobilisations.DateAndTimeReturned)
```

```python


# Merge entre les dataframes df_incidents et df_mobilisations en utilisant la colonne commune "IncidentNumber"

df = pd.merge(
    df_incidents,
    df_mobilisations,
    left_on="IncidentNumber",
    right_on="IncidentNumber",
    how="left",
)
df = df.reset_index(drop=True)

# Suppression des colonnes dupliquÃ©es
df = df.rename({"CalYear_x": "CalYear", "HourOfCall_x": "HourOfCall"}, axis=1)
df = df.drop(["CalYear_y", "HourOfCall_y"], axis=1)

print(df.shape)
display(df.head())
display(df.info())
display(df.describe())
```

```python
# Sauvegarder le DataFrame dans un fichier CSV

df.to_csv("../data/Merge.csv", sep=';')
                 
```
